### **Funcionalidades do NumPy e SciPy **

**1. Criação de Array de Inteiros Aleatórios (`np.random.randint`)**
* **Funcionalidade:** Gera um array com números inteiros aleatórios dentro de um intervalo específico (`[inicio, fim)`).
* **Aplicação no Código:** É usado para criar rapidamente um conjunto de dados de exemplo com 50 elementos, uma tarefa fundamental para testar algoritmos ou simular dados.

**2. Geração de Sequência e Indexação (`np.arange`, `.reshape`)**
* **Funcionalidade:** `np.arange(1, 17)` cria uma sequência de 1 a 16, que é então reorganizada (`reshape`) em uma matriz 4x4. A indexação `[-1]` acessa a última linha.
* **Aplicação no Código:** Demonstra a criação de uma matriz estruturada e como extrair subconjuntos específicos de dados, como a entrada mais recente de uma série.

**3. Criação de Array com Zeros e Fatiamento (`np.zeros`, slicing)**
* **Funcionalidade:** `np.zeros()` cria um array preenchido com zeros. A técnica de fatiamento (`array[4:10]`) seleciona um intervalo de índices para modificação.
* **Aplicação no Código:** Um array é inicializado e uma parte dele é "ativada" com o valor 1. Isso é muito usado para criar máscaras binárias ou definir estados iniciais em simulações.

**4. Estatística Descritiva (`np.median`, `stats.mode`)**
* **Funcionalidade:** `np.median` calcula o valor central de um conjunto de dados. `stats.mode` (da biblioteca SciPy) encontra o valor que aparece com maior frequência.
* **Aplicação no Código:** Extrai métricas estatísticas chave de um conjunto de dados aleatório para entender sua tendência central e os valores mais comuns, uma etapa inicial em qualquer análise de dados.

**5. Filtragem com Máscara Booleana**
* **Funcionalidade:** Cria uma condição (`array_5 % 3 == 0`) que gera uma máscara de `True`/`False`. Essa máscara é usada para selecionar e atribuir um novo valor (`-3`) apenas aos elementos que satisfazem a condição.
* **Aplicação no Código:** Demonstra um método eficiente e "Numpy-ônico" para substituir valores condicionalmente sem usar laços, uma técnica poderosa para limpeza e manipulação de dados.

**6. Operações por Eixo (`.sum(axis=0)`)**
* **Funcionalidade:** Realiza uma operação de agregação (neste caso, a soma) ao longo de um eixo específico da matriz. `axis=0` significa que a operação é aplicada verticalmente, em cada coluna.
* **Aplicação no Código:** Calcula a soma total de cada coluna de uma matriz 5x5, útil para agregar dados, como por exemplo, calcular o total de vendas por dia, se cada coluna representasse um dia.

**7. Coeficiente de Correlação (`np.corrcoef`)**
* **Funcionalidade:** Calcula a matriz de correlação de Pearson, que mede a força e a direção da relação linear entre duas ou mais variáveis. Os valores variam de -1 (correlação negativa perfeita) a +1 (correlação positiva perfeita).
* **Aplicação no Código:** Mede o quão linearmente relacionados são dois arrays aleatórios. É uma ferramenta estatística fundamental para investigar a associação entre variáveis.

**8. Operações Elemento a Elemento (Element-wise)**
* **Funcionalidade:** O NumPy aplica operadores matemáticos (como `** 2` para elevar ao quadrado) a cada elemento de um array individualmente.
* **Aplicação no Código:** Eleva todos os números de uma matriz 3x3 ao quadrado com uma única operação, demonstrando a eficiência da computação vetorizada para transformações de dados.

**9. Normalização de Dados (Min-Max Scaling)**
* **Funcionalidade:** Transforma os valores de um array para que eles se encaixem em uma nova escala, geralmente de 0 a 1. A fórmula `(x - min) / (max - min)` é aplicada a cada elemento.
* **Aplicação no Código:** Normaliza um array de números aleatórios. Esta é uma etapa de pré-processamento crucial para muitos algoritmos de machine learning, que funcionam melhor quando os dados de entrada estão na mesma escala.

**10. Manipulação da Diagonal (`np.eye`, `np.fill_diagonal`)**
* **Funcionalidade:** `np.eye()` cria uma matriz identidade. `np.fill_diagonal()` substitui os valores da diagonal principal por um novo conjunto de valores.
* **Aplicação no Código:** Cria uma matriz identidade e depois personaliza sua diagonal principal, uma operação comum na construção de matrizes para transformações em álgebra linear.

**11. Encontrando a Moda (`np.bincount`, `np.argmax`)**
* **Funcionalidade:** `np.bincount()` conta a frequência de cada número inteiro não negativo em um array. `np.argmax()` retorna o índice do maior valor, que, neste caso, corresponde ao número mais frequente.
* **Aplicação no Código:** Apresenta um método computacionalmente muito eficiente para encontrar o valor mais comum (moda) em um conjunto de dados de inteiros.

**12. Preenchimento de Diagonal (`np.fill_diagonal`)**
* **Funcionalidade:** Modifica a diagonal de uma matriz existente com um valor constante.
* **Aplicação no Código:** Cria uma matriz de zeros e preenche sua diagonal principal com o número 7, mostrando uma forma rápida de construir matrizes com diagonais personalizadas.

**13. Soma Cumulativa (`np.cumsum`)**
* **Funcionalidade:** Calcula a soma cumulativa dos elementos de um array. O n-ésimo elemento do resultado é a soma de todos os elementos até a n-ésima posição do array original.
* **Aplicação no Código:** Calcula a soma acumulada dos 10 primeiros números pares. É útil para acompanhar totais contínuos, como o crescimento de um investimento ao longo do tempo.

**14. Média das Diferenças entre Arrays**
* **Funcionalidade:** Realiza a subtração elemento a elemento entre dois arrays e, em seguida, calcula a média (`np.mean()`) do array resultante.
* **Aplicação no Código:** Mede a diferença média entre dois conjuntos de dados. Isso pode ser usado para, por exemplo, calcular o erro médio entre as previsões de um modelo e os valores reais.

**15. Limitação de Valores (`np.clip`)**
* **Funcionalidade:** Restringe os valores de um array a um intervalo definido, substituindo todos os valores acima do máximo pelo máximo e todos abaixo do mínimo pelo mínimo.
* **Aplicação no Código:** Substitui todos os números maiores que 50 pelo valor 50. É uma forma útil de lidar com valores extremos (outliers) ou garantir que os dados permaneçam dentro de um limite válido.

**16. Reorganização de Arrays (`.reshape`)**
* **Funcionalidade:** Modifica a forma (dimensões) de um array sem alterar seus dados.
* **Aplicação no Código:** Transforma um array unidimensional de 12 elementos em uma matriz bidimensional com 3 linhas e 4 colunas, uma operação essencial para estruturar dados.

**17. Multiplicação Elemento a Elemento com Índices**
* **Funcionalidade:** Multiplica duas matrizes de mesmo formato, elemento por elemento (não é a multiplicação de matrizes da álgebra linear).
* **Aplicação no Código:** Multiplica cada valor de uma matriz pela sua posição (índice linear). Isso pode ser usado para aplicar um "peso" a cada valor com base em sua localização.

**18. Substituição Condicional de Múltiplos**
* **Funcionalidade:** Reforça o uso de máscaras booleanas para encontrar e substituir valores que atendem a uma condição específica (ser múltiplo de 5).
* **Aplicação no Código:** Zera todos os múltiplos de 5 em um array, uma tarefa comum em pré-processamento de dados ou na implementação de certas regras lógicas.

**19. Cálculo do Ângulo entre Vetores**
* **Funcionalidade:** Utiliza uma combinação de funções de álgebra linear (`np.dot`, `np.linalg.norm`, `np.arccos`) para calcular o ângulo entre dois vetores.
* **Aplicação no Código:** Mede a "direção" relativa entre dois vetores. O cosseno do ângulo (produto escalar normalizado) é amplamente utilizado em machine learning como uma medida de similaridade de texto (similaridade de cosseno).

**20. Mínimo por Eixo (`.min(axis=1)`)**
* **Funcionalidade:** Realiza uma operação de agregação (encontrar o valor mínimo) ao longo de um eixo específico. `axis=1` significa que a operação é aplicada horizontalmente, em cada linha.
* **Aplicação no Código:** Encontra o menor valor presente em cada linha da matriz. Isso é útil para análises onde se busca o valor mínimo para cada item ou observação, como encontrar a nota mais baixa de cada aluno em uma turma.


### **Funcionalidades do SciPy **

**1. Integração Numérica (`integrate.quad`)**
* **Funcionalidade:** Calcula a integral definida de uma função de uma variável em um dado intervalo. É uma ferramenta de alta precisão para encontrar a "área sob a curva".
* **Aplicação no Código:** Calcula a integral da função $f(x) = x^2 \sin(x)$ no intervalo de 0 a $\pi$, um cálculo que seria complexo de fazer manualmente.

**2. Resolução de EDOs de 1ª Ordem (`odeint`)**
* **Funcionalidade:** `odeint` é um solver clássico para equações diferenciais ordinárias (EDOs) com um valor inicial. Ele integra a equação para encontrar a solução em vários pontos do tempo (ou espaço).
* **Aplicação no Código:** Resolve a equação $dy/dx = x - y$ com a condição inicial $y(0)=1$ e plota a trajetória da solução `y(x)`, mostrando como a variável `y` evolui.

**3. Encontrar Raízes de Polinômios (`np.roots`)**
* **Funcionalidade:** Embora seja uma função do NumPy, é frequentemente usada em contextos científicos. Ela encontra todas as raízes (complexas e reais) de um polinômio a partir de uma lista de seus coeficientes.
* **Aplicação no Código:** Resolve a equação $x^4 - 3x^3 + 2 = 0$ de forma direta e eficiente, encontrando todos os quatro valores de `x` que a satisfazem.

**4. Otimização (Maximização por Minimização)**
* **Funcionalidade:** `optimize.minimize` é projetada para encontrar o ponto de mínimo de uma função. Para maximizar uma função $f(x)$, pode-se usar um truque comum: minimizar $-f(x)$.
* **Aplicação no Código:** Encontra o ponto de máximo da função $f(x) = -x^2 + 4x + 1$ ao minimizar seu negativo, uma abordagem padrão em problemas de otimização.

**5. Transformada de Fourier (FFT)**
* **Funcionalidade:** `fft.fft` calcula a Transformada Rápida de Fourier, que decompõe um sinal do domínio do tempo em suas frequências constituintes, mostrando a amplitude de cada frequência presente.
* **Aplicação no Código:** Um sinal composto por uma onda seno e uma cosseno é criado. A FFT é calculada e o gráfico resultante mostra picos claros nas frequências correspondentes, demonstrando a análise espectral do sinal.

**6. Álgebra Linear - Inversão de Matriz (`linalg.inv`)**
* **Funcionalidade:** Calcula a matriz inversa de uma matriz quadrada, se ela existir. A matriz inversa é fundamental em muitas operações de álgebra linear.
* **Aplicação no Código:** Gera uma matriz 5x5 aleatória e calcula sua inversa, uma operação essencial para, por exemplo, resolver sistemas lineares na forma $x = A^{-1}b$.

**7. Interpolação Cúbica (`interpolate.interp1d`)**
* **Funcionalidade:** Cria uma função de interpolação que estima valores entre pontos de dados conhecidos. A opção `kind='cubic'` gera uma curva suave (um polinômio de terceiro grau por partes) que passa por todos os pontos.
* **Aplicação no Código:** A partir de 5 pontos de dados, uma curva suave é gerada para conectar todos eles, o que é útil para criar visualizações mais agradáveis ou estimar valores intermediários com maior precisão do que a interpolação linear.

**8. Álgebra Linear - Resolução de Sistemas (`linalg.solve`)**
* **Funcionalidade:** Resolve um sistema de equações lineares da forma $Ax=b$ de maneira eficiente e numericamente estável.
* **Aplicação no Código:** Encontra a solução exata `(x, y)` para o sistema de duas equações fornecido, uma tarefa fundamental em praticamente todas as áreas da ciência e engenharia.

**9. Ajuste de Curvas (`optimize.curve_fit`)**
* **Funcionalidade:** Ajusta os parâmetros de uma função definida pelo usuário para que ela se aproxime o máximo possível de um conjunto de pontos de dados.
* **Aplicação no Código:** Encontra os melhores valores para os coeficientes `a, b, c` de uma função quadrática ($ax^2+bx+c$) para que a curva resultante passe o mais perto possível dos dados observados. É uma técnica central para modelagem de dados.

**10. Integração Dupla (`integrate.dblquad`)**
* **Funcionalidade:** Estende o conceito de integração para funções de duas variáveis, calculando o volume sob a superfície definida pela função em uma região retangular.
* **Aplicação no Código:** Calcula a integral dupla da função $f(x, y) = xy + x^2$ sobre os limites especificados para `x` e `y`.

**11. Álgebra Linear - Autovalores e Autovetores (`linalg.eig`)**
* **Funcionalidade:** Calcula os autovalores e autovetores de uma matriz quadrada. Estes são valores e vetores especiais que descrevem as direções e magnitudes de "esticamento" da transformação linear representada pela matriz.
* **Aplicação no Código:** Realiza a decomposição espectral de uma matriz aleatória. Esta análise é crucial em áreas como PCA (Análise de Componentes Principais), mecânica quântica e análise de estabilidade de sistemas.

**12. Resolução de Equações Não Lineares (`optimize.fsolve`)**
* **Funcionalidade:** Encontra as raízes de uma função (ou sistema de funções) não linear, mesmo quando não há uma solução algébrica direta.
* **Aplicação no Código:** Resolve a equação transcendental $e^x = x^2$, encontrando numericamente o ponto onde as duas funções se cruzam.

**13. Álgebra Linear - Determinante (`linalg.det`)**
* **Funcionalidade:** Calcula o determinante de uma matriz quadrada, um valor escalar que fornece informações importantes sobre a matriz (por exemplo, se é invertível ou não).
* **Aplicação no Código:** Calcula o determinante de uma matriz 6x6 aleatória.

**14. Minimização de Funções Multivariadas**
* **Funcionalidade:** `optimize.minimize` pode otimizar funções com múltiplas variáveis de entrada, encontrando o vetor de parâmetros que minimiza a função.
* **Aplicação no Código:** Encontra o ponto `(x, y)` que minimiza a função $f(x, y) = (x-2)^2 + (y-3)^2$, cujo mínimo conhecido é em (2, 3).

**15. Transformada de Fourier Inversa (`fft.ifft`)**
* **Funcionalidade:** `fft.ifft` é a operação inversa da FFT. Ela reconstrói o sinal original no domínio do tempo a partir de sua representação no domínio da frequência.
* **Aplicação no Código:** Demonstra que o processo de FFT -> IFFT é reversível. Um sinal original é transformado e depois reconstruído, e a função `np.allclose` confirma que a reconstrução é praticamente idêntica ao original.

**16. Resolução de EDOs de 2ª Ordem**
* **Funcionalidade:** Solvers como `odeint` podem resolver EDOs de ordem superior, transformando-as em um sistema de EDOs de primeira ordem.
* **Aplicação no Código:** Resolve a equação do oscilador harmônico ($y'' + y = 0$) ao reescrevê-la como um sistema para $y$ e sua derivada $y'$, plotando ambas as soluções.

**17. Encontrar Raízes com `optimize.root`**
* **Funcionalidade:** `optimize.root` é um solver de raízes versátil que pode usar diferentes algoritmos.
* **Aplicação no Código:** É usado para encontrar a raiz cúbica de 27 ao formular o problema como "encontre o `x` para o qual $x^3 - 27 = 0$".

**18. Interpolação com Spline Cúbica (`interpolate.CubicSpline`)**
* **Funcionalidade:** `CubicSpline` é uma forma mais moderna e orientada a objetos de realizar interpolação cúbica. Ela cria um objeto que representa a spline e pode ser chamado como uma função.
* **Aplicação no Código:** Gera uma interpolação suave e contínua para pontos amostrados da função $y=x^2$, resultando em uma curva que não só passa pelos pontos, mas também tem derivadas contínuas.

**19. FFT de Sinal Discreto (Reprise)**
* **Funcionalidade:** Reforça o conceito da FFT como uma ferramenta para análise de frequência.
* **Aplicação no Código:** Analisa um sinal mais complexo, composto por três frequências diferentes (50, 120 e 200 Hz). O gráfico da FFT resultante mostra picos distintos em cada uma dessas três frequências, ilustrando sua capacidade de identificar múltiplos componentes em um sinal.

* **Aplicação no Código:** Resolve um sistema de três equações com três variáveis (`x, y, z`), mostrando que a mesma função usada para sistemas 2x2 funciona perfeitamente para sistemas de ordem superior.

```python
# 3 Exercicios de Pandas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Exercício 1
print("Exercício 1: Criando o DataFrame")
data = {'Produto': ['A', 'B', 'C', 'D', 'E'],
        'Preco': [120, 150, 100, 80, 200],
        'Quantidade': [10, 5, 8, 12, 6]}
df = pd.DataFrame(data)
print(df)
print("-" * 30)

# Exercício 2
print("Exercício 2: Lendo arquivo Excel (criando um arquivo de exemplo primeiro)")
# Criando um arquivo Excel
df_vendas = pd.DataFrame({
    'Vendedor': ['Ana', 'Bruno', 'Carla', 'Daniel', 'Eduarda', 'Fernanda', 'Gabriel', 'Helena', 'Igor', 'Juliana', 'Marcos'],
    'Produto': ['Celular', 'Notebook', 'Tablet', 'Fone', 'Carregador', 'Mouse', 'Teclado', 'Monitor', 'Cadeira', 'Mesa', 'Webcam'],
    'Valor': [2500, 4500, 1200, 300, 150, 200, 350, 1800, 900, 1100, 400]
})
df_vendas.to_excel("vendas.xlsx", index=False)
df_lido = pd.read_excel("vendas.xlsx")
print(df_lido.head(10))
print("-" * 30)


# Exercício 3
print("Exercício 3: Filtrando o DataFrame por preço > 100")
df_filtrado = df[df['Preco'] > 100]
print(df_filtrado)
print("-" * 30)

# Exercício 4
print("Exercício 4: Adicionando a coluna Desconto")
df['Desconto'] = df['Preco'] * 0.10
print(df)
print("-" * 30)

# Exercício 5
print("Exercício 5: Removendo linhas com valores nulos")
data_nulos = {'A': [1, 2, np.nan, 4],
              'B': [5, np.nan, 7, 8],
              'C': [9, 10, 11, 12]}
df_nulos = pd.DataFrame(data_nulos)
print("DataFrame original com valores nulos:")
print(df_nulos)
df_sem_nulos = df_nulos.dropna()
print("\nDataFrame após remover linhas com valores nulos:")
print(df_sem_nulos)
print("-" * 30)

# Exercício 6
print("Exercício 6: Ordenando o DataFrame pela coluna Produto")
df_ordenado = df.sort_values(by='Produto')
print(df_ordenado)
print("-" * 30)

# Exercício 7
print("Exercício 7: Adicionando a coluna Total")
df['Total'] = df['Preco'] * df['Quantidade']
print(df)
print("-" * 30)

# Exercício 8
print("Exercício 8: Agrupando por Categoria e calculando o total de produtos")
df['Categoria'] = ['Eletrônicos', 'Eletrônicos', 'Móveis', 'Móveis', 'Eletrônicos']
total_por_categoria = df.groupby('Categoria')['Quantidade'].sum()
print(total_por_categoria)
print("-" * 30)

# Exercício 9
print("Exercício 9: Exportando para CSV")
total_por_categoria.to_csv('resultados.csv')
print("Arquivo 'resultados.csv' criado com sucesso.")
print("-" * 30)

# Exercício 10
print("Exercício 10: Lendo CSV, renomeando colunas e exibindo as últimas linhas")
# Criando um arquivo CSV
df_csv_exemplo = pd.DataFrame({
    'col1': [1, 2, 3, 4, 5, 6, 7],
    'col2': ['A', 'B', 'C', 'D', 'E', 'F', 'G']
})
df_csv_exemplo.to_csv('exemplo.csv', index=False)
df_csv = pd.read_csv('exemplo.csv')
df_csv.rename(columns={'col1': 'ID', 'col2': 'Letra'}, inplace=True)
print(df_csv.tail(5))
print("-" * 30)

# Exercício 11
print("Exercício 11: Preenchendo valores nulos com a mediana")
data_nulos_num = {'Valores': [10, 20, np.nan, 40, 50, np.nan, 70]}
df_nulos_num = pd.DataFrame(data_nulos_num)
print("DataFrame com nulos:")
print(df_nulos_num)
mediana = df_nulos_num['Valores'].median()
df_nulos_num['Valores'].fillna(mediana, inplace=True)
print("\nDataFrame com nulos preenchidos pela mediana:")
print(df_nulos_num)
print("-" * 30)


# Exercício 12
print("Exercício 12: Verificando colunas com valores duplicados")
data_duplicados = {'A': [1, 2, 3, 1], 'B': ['x', 'y', 'z', 'w']}
df_duplicados = pd.DataFrame(data_duplicados)
for coluna in df_duplicados.columns:
    if df_duplicados[coluna].duplicated().any():
        print(f"A coluna '{coluna}' contém valores duplicados.")
    else:
        print(f"A coluna '{coluna}' não contém valores duplicados.")
print("-" * 30)

# Exercício 13
print("Exercício 13: Criando uma tabela dinâmica")
tabela_dinamica = df.pivot_table(index='Categoria', values='Quantidade', aggfunc='sum')
print(tabela_dinamica)
print("-" * 30)


# Exercício 14
print("Exercício 14: Mesclando dois DataFrames")
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Nome': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Idade': [25, 30, 22]})
df_mesclado = pd.merge(df1, df2, on='ID', how='inner')
print(df_mesclado)
print("-" * 30)

# Exercício 15
print("Exercício 15: Preenchendo dados faltantes com interpolação linear")
data_faltantes = {'Valores': [10, 20, np.nan, np.nan, 50]}
df_faltantes = pd.DataFrame(data_faltantes)
print("DataFrame com dados faltantes:")
print(df_faltantes)
df_faltantes['Valores'].interpolate(inplace=True)
print("\nDataFrame após interpolação linear:")
print(df_faltantes)
print("-" * 30)

# Exercício 16
print("Exercício 16: Criando um gráfico de barras")
df_grafico = df[['Produto', 'Quantidade']].sort_values('Quantidade', ascending=False)
ax = df_grafico.plot(kind='bar', x='Produto', y='Quantidade', legend=False, figsize=(8, 5))
ax.set_title('Quantidade por Produto')
ax.set_ylabel('Quantidade')
ax.set_xlabel('Produto')
plt.tight_layout()
plt.savefig('grafico_barras.png')
print("Gráfico de barras 'grafico_barras.png' salvo.")
print("-" * 30)

# Exercício 17
print("Exercício 17: Adicionando a coluna Imposto")
df['Imposto'] = df['Total'] * 0.05
print(df)
print("-" * 30)

# Exercício 18
print("Exercício 18: Exibindo produtos que começam com a letra 'B'")
produtos_b = df[df['Produto'].str.startswith('B')]
print(produtos_b)
print("-" * 30)

# Exercício 19
print("Exercício 19: Criando DataFrame a partir de um dicionário e calculando a média")
dic_listas = {'A': [10, 20, 30], 'B': [15, 25, 35]}
df_dic = pd.DataFrame(dic_listas)
media_colunas = df_dic.mean()
print(df_dic)
print("\nMédia de cada coluna:")
print(media_colunas)
print("-" * 30)

# Exercício 20
print("Exercício 20: Agrupando por Categoria e calculando a mediana do Preço")
mediana_preco_categoria = df.groupby('Categoria')['Preco'].median()
print(mediana_preco_categoria)
print("-" * 30)
```

### **Funcionalidades do Pandas **

**1. Criação de DataFrame a partir de um Dicionário**
* **Funcionalidade:** `pd.DataFrame()` é a função principal para criar um DataFrame, a estrutura de dados tabular (semelhante a uma planilha) do Pandas.
* **Aplicação no Código:** O código demonstra a forma mais comum de criar um DataFrame: a partir de um dicionário Python, onde as chaves se tornam os nomes das colunas e os valores (em listas) se tornam os dados dessas colunas.

**2. Leitura de Arquivos Excel (`pd.read_excel`)**
* **Funcionalidade:** O Pandas pode ler e escrever dados diretamente de e para arquivos do Microsoft Excel (`.xlsx`), o que é extremamente útil em ambientes corporativos.
* **Aplicação no Código:** O exemplo mostra o ciclo completo: primeiro, um DataFrame é criado e salvo em um arquivo "vendas.xlsx" com `to_excel()`, e depois esse mesmo arquivo é lido de volta para um novo DataFrame usando `read_excel()`. A função `.head(10)` é usada para visualizar as 10 primeiras linhas.

**3. Filtragem de Dados (Indexação Booleana)**
* **Funcionalidade:** A indexação booleana (`df[df['coluna'] > valor]`) é a principal maneira de selecionar linhas de um DataFrame que atendem a uma ou mais condições.
* **Aplicação no Código:** O código filtra o DataFrame original para manter apenas as linhas em que o valor na coluna 'Preco' é maior que 100, demonstrando uma operação fundamental de seleção de dados.

**4. Adição de Novas Colunas**
* **Funcionalidade:** Criar uma nova coluna em um DataFrame é tão simples quanto atribuir uma Série (ou um valor único) a um novo nome de coluna.
* **Aplicação no Código:** Uma nova coluna chamada 'Desconto' é criada. Seus valores são calculados aplicando uma operação matemática (multiplicação por 0.10) a uma coluna existente ('Preco'), uma tarefa comum em preparação de dados e engenharia de features.

**5. Remoção de Valores Nulos (`.dropna`)**
* **Funcionalidade:** O método `.dropna()` é usado para remover linhas (ou colunas) que contêm valores ausentes (representados como `np.nan`).
* **Aplicação no Código:** Um DataFrame com dados faltantes é criado e, em seguida, `.dropna()` é usado para eliminar todas as linhas que possuem pelo menos um valor nulo, resultando em um DataFrame "limpo".

**6. Ordenação de Dados (`.sort_values`)**
* **Funcionalidade:** Permite ordenar as linhas de um DataFrame com base nos valores de uma ou mais colunas, em ordem crescente ou decrescente.
* **Aplicação no Código:** O DataFrame é reorganizado em ordem alfabética com base nos valores da coluna 'Produto', o que é útil para visualização e relatórios.

**7. Operações Vetorizadas**
* **Funcionalidade:** O Pandas executa operações matemáticas em colunas inteiras de uma vez, sem a necessidade de laços (`for`). Isso é chamado de vetorização e é extremamente eficiente.
* **Aplicação no Código:** A coluna 'Total' é calculada multiplicando as colunas 'Preco' e 'Quantidade' diretamente. O Pandas realiza a multiplicação elemento por elemento de forma otimizada.

**8. Agrupamento e Agregação (`.groupby`)**
* **Funcionalidade:** `.groupby()` é uma das ferramentas mais poderosas do Pandas. Ele divide o DataFrame em grupos com base nos valores de uma coluna e, em seguida, aplica uma função de agregação (como soma, média, contagem) a cada grupo.
* **Aplicação no Código:** Os dados são agrupados pela coluna 'Categoria', e a soma (`.sum()`) da 'Quantidade' é calculada para cada categoria, resumindo os dados de forma eficaz.

**9. Exportação para CSV (`.to_csv`)**
* **Funcionalidade:** Salva um DataFrame em um arquivo de formato CSV (Valores Separados por Vírgula), um formato de texto simples e universal para troca de dados.
* **Aplicação no Código:** O resultado do agrupamento do exercício anterior é salvo no arquivo `resultados.csv`, demonstrando como exportar dados processados.

**10. Leitura de CSV e Renomeação de Colunas (`.read_csv`, `.rename`, `.tail`)**
* **Funcionalidade:** Demonstra o fluxo de trabalho de ler um arquivo CSV, renomear as colunas para nomes mais descritivos com `.rename()` e visualizar as últimas linhas com `.tail()`.
* **Aplicação no Código:** Um arquivo CSV de exemplo é lido, suas colunas 'col1' e 'col2' são renomeadas para 'ID' e 'Letra', e as últimas 5 linhas do DataFrame resultante são exibidas.

**11. Preenchimento de Valores Nulos (`.fillna`)**
* **Funcionalidade:** Em vez de remover dados ausentes, `.fillna()` permite preenchê-los com um valor específico, como a média, mediana ou moda da coluna.
* **Aplicação no Código:** Os valores `NaN` na coluna 'Valores' são substituídos pela mediana da própria coluna, uma estratégia comum para tratar dados faltantes sem remover linhas inteiras.

**12. Verificação de Duplicados (`.duplicated`)**
* **Funcionalidade:** O método `.duplicated()` retorna uma Série booleana indicando se cada linha (ou valor, se aplicado a uma Série) é uma duplicata de uma anterior.
* **Aplicação no Código:** O código itera sobre as colunas e usa `.duplicated().any()` para verificar se existe algum valor duplicado dentro de cada coluna, uma etapa importante na verificação da qualidade dos dados.

**13. Tabelas Dinâmicas (`.pivot_table`)**
* **Funcionalidade:** Cria um resumo de dados no estilo de uma tabela dinâmica de planilha, permitindo reorganizar e agregar dados de forma flexível.
* **Aplicação no Código:** Uma tabela dinâmica é criada para mostrar a soma (`aggfunc='sum'`) da 'Quantidade' para cada 'Categoria', apresentando os mesmos resultados do `.groupby()` de uma forma diferente.

**14. Mesclagem de DataFrames (`pd.merge`)**
* **Funcionalidade:** Combina dois DataFrames com base em uma ou mais colunas em comum, semelhante a um `JOIN` em bancos de dados SQL.
* **Aplicação no Código:** Dois DataFrames (`df1` e `df2`) são mesclados usando a coluna 'ID' como chave. O `how='inner'` garante que apenas as linhas com IDs presentes em ambos os DataFrames sejam mantidas no resultado.

**15. Interpolação de Dados Faltantes (`.interpolate`)**
* **Funcionalidade:** É um método mais avançado para preencher valores `NaN` em dados numéricos. Ele estima os valores ausentes com base nos valores vizinhos (por padrão, usando uma interpolação linear).
* **Aplicação no Código:** Os dois valores `NaN` entre 20 e 50 são preenchidos com 30 e 40, respectivamente, como se estivessem em uma linha reta entre os pontos conhecidos.

**16. Plotagem Direta do DataFrame (`.plot`)**
* **Funcionalidade:** O Pandas integra-se diretamente com o Matplotlib, permitindo criar gráficos rapidamente usando o método `.plot()` em um DataFrame ou Série.
* **Aplicação no Código:** Um gráfico de barras da 'Quantidade' por 'Produto' é gerado diretamente do DataFrame, simplificando muito o processo de visualização de dados.

**17. Adição de Outra Coluna Calculada**
* **Funcionalidade:** Reforça a facilidade com que novas colunas podem ser criadas a partir de operações em colunas existentes.
* **Aplicação no Código:** Uma coluna 'Imposto' é adicionada, calculada como 5% do valor da coluna 'Total', demonstrando novamente as operações vetorizadas.

**18. Operações com Strings (`.str`)**
* **Funcionalidade:** O acessador `.str` permite aplicar uma vasta gama de funções de manipulação de texto a uma coluna inteira de forma vetorizada.
* **Aplicação no Código:** `.str.startswith('B')` é usado para filtrar e exibir apenas os produtos cujo nome começa com a letra 'B', uma poderosa ferramenta para trabalhar com dados textuais.

**19. Agregação em DataFrame (`.mean`)**
* **Funcionalidade:** Funções de agregação como `.mean()`, `.sum()`, `.max()`, etc., podem ser aplicadas a um DataFrame inteiro para calcular a estatística para cada coluna numérica.
* **Aplicação no Código:** Um DataFrame é criado e `.mean()` é chamado para calcular a média de cada coluna ('A' e 'B') de uma só vez.

**20. Agrupamento com Agregação Diferente**
* **Funcionalidade:** Demonstra a flexibilidade do `.groupby()`, que pode ser combinado com qualquer função de agregação.
* **Aplicação no Código:** Os dados são novamente agrupados por 'Categoria', mas desta vez a mediana (`.median()`) do 'Preco' é calculada para cada grupo, mostrando uma análise diferente da realizada no exercício 8.

### **Funcionalidades do Matplotlib**

**1. Gráfico de Linha Básico (`plt.plot`)**
* **Funcionalidade:** A função `plot` é usada para criar gráficos de linha, que são ideais para visualizar a relação entre duas variáveis contínuas.
* **Aplicação no Código:** O código plota a função matemática $y = x^3$, mostrando claramente como os valores de `y` mudam em relação aos valores de `x` em uma linha contínua.

**2. Gráfico de Dispersão com Mapa de Cores (`plt.scatter`, `cmap`)**
* **Funcionalidade:** `scatter` cria um gráfico de dispersão, mostrando a relação entre duas variáveis. O argumento `c` junto com `cmap` (mapa de cores) permite que uma terceira variável seja representada pela cor de cada ponto.
* **Aplicação no Código:** Visualiza a relação entre `x` e `y` aleatórios. A cor de cada ponto é determinada pelo seu valor `y`, criando um gradiente de cor que adiciona uma dimensão extra de informação visual.

**3. Histograma (`plt.hist`)**
* **Funcionalidade:** Um histograma agrupa dados em "caixas" (bins) e mostra a frequência de ocorrência de dados em cada caixa. É perfeito para visualizar a distribuição de uma única variável.
* **Aplicação no Código:** O código gera 500 números inteiros aleatórios e usa o histograma para mostrar quais faixas de números apareceram com mais ou menos frequência.

**4. Gráfico de Barras (`plt.bar`)**
* **Funcionalidade:** O gráfico de barras é usado para comparar valores numéricos entre diferentes categorias discretas.
* **Aplicação no Código:** Compara os valores de cinco categorias ('A', 'B', 'C', 'D', 'E'), onde a altura de cada barra representa o valor correspondente. Cores diferentes são usadas para distinguir visualmente cada categoria.

**5. Gráfico de Pizza (`plt.pie`)**
* **Funcionalidade:** Este gráfico representa a composição de um todo, mostrando a proporção de cada categoria como uma "fatia" do total.
* **Aplicação no Código:** Mostra a participação percentual de diferentes setores de mercado. O argumento `autopct` é usado para exibir automaticamente o valor percentual em cada fatia.

**6. Subplots (`plt.subplots`)**
* **Funcionalidade:** Permite criar múltiplos gráficos (eixos) dentro de uma única figura, facilitando a comparação direta entre diferentes visualizações.
* **Aplicação no Código:** Cria uma figura com dois gráficos lado a lado: um para a função exponencial ($y=e^x$) e outro para a função logarítmica ($y=\log(x)$), permitindo uma comparação visual imediata.

**7. Gráfico de Dispersão com Tamanho Variável (`plt.scatter`, `s`)**
* **Funcionalidade:** Similar a um gráfico de dispersão padrão, mas o argumento `s` (size) permite que uma terceira variável seja representada pelo tamanho de cada ponto.
* **Aplicação no Código:** Cria um "gráfico de bolhas" onde o tamanho de cada ponto é proporcional ao seu valor `y`, adicionando outra camada de informação visual à relação entre as variáveis.

**8. Gráfico com Área Preenchida (`plt.fill_between`)**
* **Funcionalidade:** Preenche a área entre duas curvas ou em torno de uma linha. É comumente usado para indicar um intervalo de confiança, uma margem de erro ou uma faixa de variação.
* **Aplicação no Código:** Desenha a função seno e preenche uma área cinza e semitransparente ao redor dela, simulando visualmente um "intervalo de confiança".

**9. Grade Personalizada (`plt.grid`)**
* **Funcionalidade:** Adiciona uma grade ao fundo do gráfico para facilitar a leitura dos valores. Seus parâmetros permitem personalizar o estilo, a espessura e a cor das linhas.
* **Aplicação no Código:** Adiciona uma grade com linhas tracejadas (`--`), finas e cinzas, tornando o gráfico visualmente mais limpo e preciso.

**10. Múltiplas Linhas no Mesmo Gráfico**
* **Funcionalidade:** Chamar a função `plt.plot` várias vezes para o mesmo eixo desenha múltiplas linhas no mesmo gráfico. Estilos de linha e cores diferentes podem ser usados para diferenciá-las.
* **Aplicação no Código:** Plota três funções trigonométricas (seno, cosseno e a soma delas) juntas, usando cores e estilos de linha distintos (tracejado, pontilhado, sólido) para que cada uma seja facilmente identificável.

**11. Gráfico de Barras Empilhadas (`plt.bar`, `bottom`)**
* **Funcionalidade:** As barras empilhadas mostram a composição de cada categoria principal. Cada segmento da barra representa uma subcategoria, e a altura total da barra representa o total da categoria.
* **Aplicação no Código:** Para cada grupo, três séries de valores são empilhadas uma sobre a outra. O parâmetro `bottom` é a chave, pois define onde cada nova série de barras deve começar a ser desenhada.

**12. Limites dos Eixos (`plt.xlim`, `plt.ylim`)**
* **Funcionalidade:** Permite definir manualmente o intervalo visível dos eixos X e Y, funcionando como um "zoom" para focar em uma área de interesse específica do gráfico.
* **Aplicação no Código:** Plota uma função quadrática completa, mas ajusta os eixos para mostrar apenas a região de $x$ entre -5 e 5 e de $y$ entre 0 e 50.

**13. Gráfico de Dispersão 3D (`Axes3D`)**
* **Funcionalidade:** Estende a visualização para três dimensões, permitindo plotar pontos de dados que possuem três variáveis numéricas (X, Y e Z).
* **Aplicação no Código:** Cria um gráfico de dispersão tridimensional, mostrando a distribuição de pontos aleatórios em um espaço 3D.

**14. Gráfico de Barras Horizontal (`plt.barh`)**
* **Funcionalidade:** Uma variação do gráfico de barras padrão onde as barras são desenhadas horizontalmente. É especialmente útil quando os nomes das categorias são longos.
* **Aplicação no Código:** Compara valores de quatro categorias, mas com as barras crescendo da esquerda para a direita, o que pode facilitar a leitura dos rótulos no eixo Y.

**15. Gráfico de Função Inversa (`plt.plot`)**
* **Funcionalidade:** Demonstra novamente a versatilidade da função `plot` para visualizar qualquer relação matemática.
* **Aplicação no Código:** Plota a função $y = 1/x$, mostrando como `y` decresce rapidamente à medida que `x` aumenta.

**16. Anotações de Texto (`plt.text`)**
* **Funcionalidade:** Adiciona um texto personalizado em qualquer coordenada (x, y) dentro do gráfico. É útil para destacar pontos específicos ou adicionar informações contextuais.
* **Aplicação no Código:** Adiciona o rótulo "Ponto Importante" diretamente no gráfico, próximo a um ponto específico da curva da raiz quadrada.

**17. Salvando em Diferentes Formatos (`plt.savefig`)**
* **Funcionalidade:** Salva a figura gerada em um arquivo. O formato do arquivo (PNG, JPG, PDF, SVG) é determinado pela extensão fornecida no nome do arquivo.
* **Aplicação no Código:** Salva o gráfico da função cosseno em formato **PDF**, que é um formato vetorial ideal para publicações de alta qualidade, pois não perde resolução ao ser ampliado.

**18. Coloração Condicional de Barras**
* **Funcionalidade:** Permite definir a cor de cada elemento do gráfico (neste caso, cada barra) com base em uma condição. Uma linha de referência horizontal (`axhline`) também pode ser adicionada.
* **Aplicação no Código:** Cria um gráfico de barras onde os valores negativos são coloridos de vermelho e os positivos de azul. A linha horizontal em $y=0$ serve como uma referência clara para separar os valores.

**19. Legendas (`plt.legend`)**
* **Funcionalidade:** Adiciona uma legenda ao gráfico, que explica o que cada cor, estilo de linha ou símbolo representa. É essencial para gráficos com múltiplas séries de dados.
* **Aplicação no Código:** Plota duas curvas ($x^2$ e $x^3$) e adiciona uma legenda que identifica claramente qual linha corresponde a cada função.

**20. Ticks Personalizados (`plt.xticks`, `plt.yticks`)**
* **Funcionalidade:** Permite controlar manualmente a posição e os rótulos dos marcadores (ticks) nos eixos X e Y.
* **Aplicação no Código:** Define que os marcadores do eixo X devem aparecer a cada 1 unidade (de 0 a 10) e os do eixo Y a cada 10 unidades (de 0 a 100), tornando a grade e a leitura dos valores mais organizadas.
