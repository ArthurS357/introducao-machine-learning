```python
#Calculadora de Medias

def calcular_media():

    numeros = []

    print("--- Calculadora de Média Aritmética ---")
    print("Digite os números um por um. Pressione Enter sem digitar nada para finalizar.")

    while True:
        entrada_usuario = input("Digite um número: ")

        # Verifica se o usuário pressionou Enter sem digitar nada para sair do loop.
        if entrada_usuario == "":
            break

        try:
            numero = float(entrada_usuario)
            numeros.append(numero)
        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números.")

    # Após o loop, verifica se a lista de números não está vazia.
    if numeros:
        # Soma de todos os números na lista.
        soma_total = sum(numeros)
        # Quantidade de números na lista.
        quantidade_numeros = len(numeros)
        # Média aritmética.
        media = soma_total / quantidade_numeros

        print("\n--- Resultado ---")
        print(f"Os números inseridos foram: {numeros}")
        print(f"A média aritmética dos números é: {media:.2f}")
    else:
        # Se nenhum número foi inserido, exibe uma mensagem informativa.
        print("\nNenhum número foi inserido. Não é possível calcular a média.")

if __name__ == "__main__":
    calcular_media()
```

```python
# Contador de caracteres


def contar_caracteres():
    string_usuario = input("Digite uma frase ou palavra: ")

    # Remove todos os espaços da string
    string_sem_espacos = string_usuario.replace(" ", "")
    frequencia = {}

    # Itera sobre cada caractere da string sem espaços
    for caractere in string_sem_espacos:
        # Se o caractere já está no dicionário, incrementa sua contagem.
        # Se não, adiciona ao dicionário com a contagem inicial de 1.
        frequencia[caractere] = frequencia.get(caractere, 0) + 1

    print("\n--- Frequência de Caracteres ---")

    # Exibe o resultado
    if frequencia:
        for caractere, contagem in frequencia.items():
            print(f"'{caractere}': {contagem} vez(es)")
    else:
        print("Nenhum caractere foi inserido.")

if __name__ == "__main__":
    contar_caracteres()

```

```python
import string

def verificar_palindromo():
    frase_usuario = input("Digite uma palavra ou frase para verificar: ")

    # 1. Converte toda a string para minúsculas
    frase_lower = frase_usuario.lower()

    # 2. Remove espaços e pontuações usando str.translate
    translator = str.maketrans('', '', string.punctuation + ' ')
    frase_limpa = frase_lower.translate(translator)

    # 3. Inverte a string limpa
    frase_invertida = frase_limpa[::-1]

    print("\n--- Verificando ---")
    print(f"Texto original tratado: '{frase_limpa}'")
    print(f"Texto invertido: '{frase_invertida}'")

    # 4. Compara a string limpa com a sua versão invertida
    if frase_limpa == frase_invertida:
        print(f"\nResultado: Sim, '{frase_usuario}' é um palíndromo!")
    else:
        print(f"\nResultado: Não, '{frase_usuario}' não é um palíndromo.")
if __name__ == "__main__":
    verificar_palindromo()
```

```python
#Conversor de temperatura

def celsius_para_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_para_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

print("Escolha a conversão:")
print("1. Celsius para Fahrenheit")
print("2. Fahrenheit para Celsius")

opcao = input("Digite 1 ou 2: ")

if opcao == '1':
    celsius = float(input("Digite a temperatura em Celsius: "))
    fahrenheit = celsius_para_fahrenheit(celsius)
    print(f"{celsius}°C é equivalente a {fahrenheit:.2f}°F.")
elif opcao == '2':
    fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
    celsius = fahrenheit_para_celsius(fahrenheit)
    print(f"{fahrenheit}°F é equivalente a {celsius:.2f}°C.")
else:
    print("Opção inválida. Por favor, escolha 1 ou 2.")
```

```python
#Gerador de Tabuada
def gerar_tabuada():

    print("--- Gerador de Tabuada ---")

    try:
        # Solicita ao usuário que digite um número e tenta convertê-lo para inteiro.
        numero = int(input("Digite um número inteiro para ver a tabuada: "))

        print(f"\n--- Tabuada do {numero} ---")

        # O laço 'for' irá repetir 10 vezes, com a variável 'multiplicador' indo de 1 a 10.
        # A função range(1, 11) gera uma sequência de números começando em 1 e terminando em 10.
        for multiplicador in range(1, 11):
            resultado = numero * multiplicador

            # Exibe a linha da tabuada de forma formatada.
            print(f"{numero} x {multiplicador} = {resultado}")

    except ValueError:
        print("\nErro: Entrada inválida. Por favor, digite um número inteiro.")
if __name__ == "__main__":
    gerar_tabuada()
```

```python
# Simulador de Lançamento de Dados

import random

def simulador_de_dado():
    print("--- Simulador de Lançamento de Dados ---")
    try:
        vezes = int(input("Quantas vezes você deseja lançar o dado? "))
        if vezes <= 0:
            print("Por favor, digite um número positivo.")
            return

        print(f"\nLançando o dado {vezes} vez(es):")
        for i in range(1, vezes + 1):
            resultado = random.randint(1, 6)
            print(f"Lançamento {i}: {resultado}")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if __name__ == "__main__":
    simulador_de_dado()
```

```python
# Ordenação de Lista
def ordenar_lista():
    print("--- Ordenação de Lista ---")
    print("Digite os números um por um. Pressione Enter sem digitar nada para finalizar.")
    numeros = []

    while True:
        entrada = input("Digite um número: ")
        if entrada == "":
            break
        try:
            numeros.append(float(entrada))
        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números.")

    if not numeros:
        print("Nenhum número foi inserido.")
        return

    print("\nLista original:", numeros)

    numeros.sort()

    print("Lista ordenada:", numeros)

if __name__ == "__main__":
    ordenar_lista()
```

```python
# Verificador de numeros primos

def verificar_primo():
    print("--- Verificador de Números Primos ---")
    try:
        numero = int(input("Digite um número inteiro maior que 1: "))

        if numero <= 1:
            print("O número deve ser maior que 1.")
            return

        divisores = []
        for i in range(1, numero + 1):
            if numero % i == 0:
                divisores.append(i)

        print("\nDivisores:", divisores)

        if len(divisores) == 2:
            print(f"O número {numero} é primo.")
        else:
            print(f"O número {numero} não é primo.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if __name__ == "__main__":
    verificar_primo()
```

```python
# Sistema de Longin Simples
def sistema_login():
    print("--- Sistema de Login ---")
    credenciais = {}

    while True:
        print("\n1. Cadastrar")
        print("2. Login")
        print("3. Sair")

        escolha = input("Escolha uma opção: ")

        if escolha == '1':
            usuario = input("Digite um nome de usuário: ")
            if usuario in credenciais:
                print("Usuário já existe. Tente outro nome.")
            else:
                senha = input("Digite uma senha: ")
                credenciais[usuario] = senha
                print("Usuário cadastrado com sucesso!")

        elif escolha == '2':
            usuario = input("Usuário: ")
            senha = input("Senha: ")
            if credenciais.get(usuario) == senha:
                print("Login bem-sucedido!")
            else:
                print("Usuário ou senha inválidos.")

        elif escolha == '3':
            print("Saindo do sistema.")
            break

        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    sistema_login()
```

```python
# Calculadora de Fatorial
def calcular_fatorial():
    print("--- Calculadora de Fatorial ---")
    try:
        numero = int(input("Digite um número inteiro não negativo: "))

        if numero < 0:
            print("Não é possível calcular o fatorial de um número negativo.")
            return

        fatorial = 1

        if numero > 0:
            for i in range(1, numero + 1):
                fatorial *= i

        print(f"O fatorial de {numero} é {fatorial}")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if __name__ == "__main__":
    calcular_fatorial()
```

```python
# Conversor de Moedas Simples

def conversor_moeda(valor, moeda_destino):
    taxa_cambio = {'dolar': 5.25, 'euro': 5.80}
    if moeda_destino in taxa_cambio:
        return valor / taxa_cambio[moeda_destino]
    return None

def main():
    try:
        valor_reais = float(input("Insira o valor em reais: "))
        moeda = input("Escolha a moeda de destino (dolar/euro): ").lower()
        resultado = conversor_moeda(valor_reais, moeda)

        if resultado is not None:
            print(f"O valor em {moeda.capitalize()} é: {resultado:.2f}")
        else:
            print("Moeda de destino inválida.")
    except ValueError:
        print("Entrada inválida. Por favor, insira um número para o valor.")

if __name__ == "__main__":
    main()
```

```python
#Gerador de numeros Aleatorios

import random

def gerador_numeros_aleatorios(n, limite_inferior, limite_superior):
    return [random.randint(limite_inferior, limite_superior) for _ in range(n)]

def main():
    try:
        n = int(input("Quantos números aleatórios você deseja gerar? "))
        limite_inferior = int(input("Insira o limite inferior: "))
        limite_superior = int(input("Insira o limite superior: "))
        numeros = gerador_numeros_aleatorios(n, limite_inferior, limite_superior)
        print("Números gerados:", numeros)
    except ValueError:
        print("Entrada inválida. Por favor, insira apenas números inteiros.")

if __name__ == "__main__":
    main()
```

```python
# Calculadora de Média Ponderada

def media_ponderada(notas, pesos):
    soma_ponderada = sum(n * p for n, p in zip(notas, pesos))
    if sum(pesos) == 0:
        return 0
    return soma_ponderada / sum(pesos)

def main():
    try:
        notas = list(map(float, input("Insira as notas separadas por espaço: ").split()))
        pesos = list(map(float, input("Insira os pesos correspondentes separadas por espaço: ").split()))

        if len(notas) != len(pesos):
            print("Erro: O número de notas e pesos deve ser o mesmo.")
            return

        resultado = media_ponderada(notas, pesos)
        print(f"Média ponderada: {resultado:.2f}")
    except ValueError:
        print("Entrada inválida. Certifique-se de inserir apenas números.")

if __name__ == "__main__":
    main()
```

```python
# Cálculo de Raiz Quadrada
import math

def calcular_raiz():
    try:
        numero = float(input("Digite um número para calcular a raiz quadrada: "))
        if numero >= 0:
            raiz = math.sqrt(numero)
            print(f"A raiz quadrada de {numero} é: {raiz:.2f}")
        else:
            print("Não é possível calcular a raiz quadrada de um número negativo.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")

if __name__ == "__main__":
    calcular_raiz()
```

```python
# Simulador de Caixa Eletrônico
def simulador_caixa(valor):
    notas = [100, 50, 20, 10]
    resultado = {}
    valor_restante = valor

    if valor_restante % min(notas) != 0:
        return None

    for nota in notas:
        quantidade = valor_restante // nota
        if quantidade > 0:
            resultado[nota] = int(quantidade)
            valor_restante %= nota
    return resultado

def main():
    try:
        valor_saque = int(input("Insira o valor que deseja sacar: "))
        resultado = simulador_caixa(valor_saque)

        if resultado:
            print("Notas a serem entregues:")
            for nota, quantidade in resultado.items():
                print(f"R${nota}: {quantidade} nota(s)")
        else:
            print("Não é possível sacar esse valor com as notas disponíveis (R$10, R$20, R$50, R$100).")
    except ValueError:
        print("Entrada inválida. Por favor, insira um número inteiro para o valor.")

if __name__ == "__main__":
    main()
```

```python
# Verificador de Ano Bissexto
def ano_bissexto(ano):
    return (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)

def main():
    try:
        ano_usuario = int(input("Digite um ano: "))
        if ano_bissexto(ano_usuario):
            print(f"{ano_usuario} é um ano bissexto.")
        else:
            print(f"{ano_usuario} não é um ano bissexto.")
    except ValueError:
        print("Entrada inválida. Por favor, insira um ano válido.")

if __name__ == "__main__":
    main()
```

```python
#Gerenciador de contatos simples
contatos = {}

def adicionar_contato(nome, telefone):
    contatos[nome] = telefone
    print(f"Contato {nome} adicionado.")

def remover_contato(nome):
    if nome in contatos:
        del contatos[nome]
        print(f"Contato {nome} removido.")
    else:
        print(f"Contato {nome} não encontrado.")

def listar_contatos():
    if contatos:
        print("Lista de Contatos:")
        for nome, telefone in contatos.items():
            print(f"{nome}: {telefone}")
    else:
        print("Nenhum contato na lista.")

def main():
    while True:
        acao = input("Escolha uma ação (adicionar/remover/listar/sair): ").lower()
        if acao == "adicionar":
            nome = input("Nome: ")
            telefone = input("Telefone: ")
            adicionar_contato(nome, telefone)
        elif acao == "remover":
            nome = input("Nome do contato para remover: ")
            remover_contato(nome)
        elif acao == "listar":
            listar_contatos()
        elif acao == "sair":
            print("Saindo do gerenciador de contatos.")
            break
        else:
            print("Ação inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```

```python
#Calculadora de descontos

def calcular_desconto(preco_original, percentual_desconto):
    if not 0 <= percentual_desconto <= 100:
        print("Erro: O percentual de desconto deve estar entre 0 e 100.")
        return None
    preco_final = preco_original * (1 - percentual_desconto / 100)
    return preco_final

def main():
    try:
        preco_original = float(input("Insira o preço original do produto: R$ "))
        percentual_desconto = float(input("Insira a porcentagem de desconto: "))
        preco_final = calcular_desconto(preco_original, percentual_desconto)
        if preco_final is not None:
            print(f"O preço final após o desconto é: R$ {preco_final:.2f}")
    except ValueError:
        print("Entrada inválida. Por favor, insira apenas números.")

if __name__ == "__main__":
    main()
```

```python
# Análise de Dados Simples com Pandas e Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def main():
    # Carregar o conjunto de dados do Titanic
    try:
        titanic = sns.load_dataset('titanic')

        # Calcular estatísticas descritivas
        estatisticas = titanic.describe()
        print("Estatísticas Descritivas:")
        print(estatisticas)

        # Calcular média e mediana da idade
        media_idade = titanic['age'].mean()
        mediana_idade = titanic['age'].median()
        print(f"\nMédia da idade: {media_idade:.2f}")
        print(f"Mediana da idade: {mediana_idade:.2f}")

        # Criar um histograma da idade
        plt.figure(figsize=(10, 6))
        sns.histplot(titanic['age'].dropna(), bins=30, kde=True)
        plt.title('Distribuição da Idade dos Passageiros do Titanic')
        plt.xlabel('Idade')
        plt.ylabel('Frequência')
        plt.show()

        # Criar um gráfico de barras para a contagem de sobreviventes
        plt.figure(figsize=(8, 5))
        sns.countplot(x='survived', data=titanic)
        plt.title('Contagem de Sobreviventes do Titanic')
        plt.xlabel('Sobreviveu (0 = Não, 1 = Sim)')
        plt.ylabel('Contagem')
        plt.xticks(ticks=[0, 1], labels=['Não', 'Sim'])
        plt.show()

    except Exception as e:
        print(f"Ocorreu um erro: {e}")
        print("Verifique se as bibliotecas pandas, matplotlib e seaborn estão instaladas e se o conjunto de dados 'titanic' pode ser carregado.")

if __name__ == "__main__":
    main()
```
